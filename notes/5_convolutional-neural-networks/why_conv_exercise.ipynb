1.这一章节解释了：为什么处理图片时，不能再像处理表格那样用普通全连接网络，而要发明卷积神经网络（CNN）

图片与表格不同，图片有非常强的空间结构，与像素有关。卷积能够尊重图片的规律。



2.为什么多层感知器（全连接）不适合图片

在表格中，特征之间存在联系但并不是空间关系，而图片是由像素组成的，参数很多，训练慢数据多容易过拟合

多层感知器会将图像看作普通向量处理，而CNN会利用图像本身的结构。

3.不变性

物体的位置变了，但物体的身份不该变。

空间不变性：同样的“识别规则”，应该能在图像的不同位置重复使用。卷积核就像一个小探测器，它不关心自己现在在图的哪里，只关心当前这块区域像不像某种模式。

全连接层：不同位置用不同参数
卷积层：不同位置共享同一套参数

4.CNN 的设计思想归纳成两个原则：空间不变性和局部性

局部性：图片前面几层，不需要一上来就看全图；先看局部就够了。

二者决定了CNN的结构：平移不变性决定参数共享，局部性决定小窗口连接而不是全连接

5.通道：为什么图片不只是二维，而是三维

图象是二维的包括高和宽，但真实的彩色图片还有第三维：红绿蓝（RGB），所以一张图片是三位张量，每个像素不再是一个数，而是一个小向量

输入通道：输入图片原本会有几层（灰度一通道，RGB图三通道），输出通道即卷积层想学厨多少种不同特征，64个输出通道就是要学64种不同的局部模式

需要多个通道是因为一个位置上的图像信息通常不止一个方面，对同一位置我们要有多方面描述。

卷积层不是只产生一张图，而是会产生很多张特征图。每一张特征图都像是在看图像的某一个方面，实现同时提取多种特征。
